```python
import numpy as np
import pandas as pd
import random

init = [
    [2, 0, 0, 0, 2, 4, 2, 0],
    [0, 4, 0, 6, 0, 0, 0, 0],
    [3, 0, 4, 0, 2, 0, 1, 0],
    [0, 5, 0, 5, 0, 0, 0, 0],
    [1, 0, 4, 0, 3, 2, 0, 0],
    [2, 0, 0, 0, 2, 5, 1, 0],
    [4, 0, 3, 0, 0, 3, 0, 0],
    [0, 3, 0, 1, 0, 3, 3, 0],
]

for row in range(len(init)):
    for col in range(len(init)):
        init[row][col] /= 10

P = np.array(init)

states = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
def walk(P, steps, states, start_state=0):
    cur_state = start_state
    if start_state == 0:
        cur_state = random.choice(states)
        
    path = [cur_state]
    
    for i in range(steps - 1):
        weights = P[cur_state-1]
        cur_state = random.choices(states, weights)[0]
        path.append(cur_state)

    return path
```

### Модель траектории длинны 10

```python
path = walk(P, 10, states)
print('Траектория длинны 10:')
print(path)
```

### Модель траектории длинны 50

```python
path = walk(P, 50, states)
print('Траектория длинны 50:')
print(path)
```

### Модель траектории длинны 100

```python
path = walk(P, 100, states)
print('Траектория длинны 100:')
print(path)
```

### Модель траектории длинны 1000

```python
path = walk(P, 1000, states)
print('Траектория длинны 1000:')
print(path)
```

### Процент времени нахождения в каждом из состояний

Посчитаем путь длинны 10_000_000 для трех начальных состояний - по одному из каждого класса E0, E1, E2

```python
path_for_E_0 = walk(P, 10_000_000, states, 8)
path_for_E_1 = walk(P, 10_000_000, states, 1)
path_for_E_2 = walk(P, 10_000_000, states, 2)

# E0
count_8 = 0
total_E_0 = len(path_for_E_0)
for i in range(total_E_0):
    if path_for_E_0[i] == 8:
        count_8 += 1
        
prob_8 = count_8 / total_E_0

# E1
count_1 = 0
count_3 = 0
count_5 = 0
count_6 = 0
count_7 = 0
total_E_1 = len(path_for_E_1)
for i in range(total_E_1):
    match path_for_E_1[i]:
        case 1:
            count_1 += 1
        case 3:
            count_3 += 1
        case 5:
            count_5 += 1
        case 6:
            count_6 += 1
        case 7:
            count_7 += 1

prob_1 = count_1 / total_E_1
prob_3 = count_3 / total_E_1
prob_5 = count_5 / total_E_1
prob_6 = count_6 / total_E_1
prob_7 = count_7 / total_E_1

# E2
count_2 = 0
count_4 = 0
total_E_2 = len(path_for_E_2)
for i in range(total_E_2):
    match path_for_E_2[i]:
        case 2:
            count_2 += 1
        case 4:
            count_4 += 1

prob_2 = count_2 / total_E_2
prob_4 = count_4 / total_E_2


print('состояние - процент нахождения (вероятность) - финальная фероятность')
print(f'1 - {prob_1} - 0.2')
print(f'2 - {prob_2} - 0.5')
print(f'3 - {prob_3} - 0.2')
print(f'4 - {prob_4} - 0.5')
print(f'5 - {prob_5} - 0.2')
print(f'6 - {prob_6} - 0.2')
print(f'7 - {prob_7} - 0.2')
print(f'8 - {prob_8} - 0')
```

В целом вероятности довольно близки к финальным.
